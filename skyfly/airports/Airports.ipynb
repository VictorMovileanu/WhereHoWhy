```python
import pandas as pd
```

```python
pd.read_json('airports.json')
```

```python
df = pd.read_json('airports.json')
```

```python
df = df.transpose()
```

```python
df = df[df['iata'] != '']
```

```python
df = df[df['city'] != '']
```

```python
df = df[['city', 'country', 'iata']]
```

```python
country_codes = pd.read_csv('country_codes.csv', index_col=0)
```

```python
country_by_continent = pd.read_csv('coutry_codes_by_continent.csv', index_col=0, na_filter=False)
```

```python
countries_and_continents = country_codes.join(country_by_continent)
```

```python
continents = pd.DataFrame({'continent names':['Europe', 'Asia', 'North America', 'Africa', 'South America', 'Oceania', 'Antarctica']}, index=['EU','AS','NA','AF','SA','OC','AN'])
```

```python
countries_and_continents = pd.merge(countries_and_continents, continents, left_on='continent code', right_index=True, how='left')
```

```python
result = pd.merge(df, countries_and_continents, left_on='country', right_index=True)
```

```python
with open('iata_codes.json', 'w') as f:
    f.write(str(list(result['iata'])))
```
